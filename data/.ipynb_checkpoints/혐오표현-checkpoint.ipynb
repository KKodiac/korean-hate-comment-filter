```python
new_coment = []
token = []

for coment in coments:
    token = okt.morphs(coment)
    #poss = okt.pos(coment)
    
    temp_token = []
    for word in token:
        if word in stopwords:
            temp_token.append('*')
        if word not in stopwords:
            temp_token.append(word)
    
    new_coment.append(temp_token)
```

```python
complete = []
tem_complete = []

for coment in new_coment:
    new_com = list(set(coment))
    
    tem_complete.append("*".join(new_com))
```
